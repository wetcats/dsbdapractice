# Pyplot tutorial

An introduction to the pyplot interface.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("heart.csv")
```

```python
print(df)
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
sns.countplot(x="Sex",data=df,palette="bwr")
plt.title("Sex count")
```

```python
sns.countplot(x="HeartDisease", data = df, color= "blue")
```

```python
pd.crosstab(df.Age,df.HeartDisease).plot(kind="bar",figsize=(20,6))
```

```python
dp = df.HeartDisease
ns = dp.groupby(pd.cut(df['RestingBP'],[100,120,140,160])).count()
plt.pie(ns,labels=['low','mid','hight'])
```
